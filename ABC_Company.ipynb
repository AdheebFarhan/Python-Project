```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_excel(r"C:\Users\user\Downloads\myexcel.xlsx")
df
```

Data preprocessing 

```python
# Remove number and college columns
df.drop(['Number','College'], axis=1, inplace=True)
df
```

```python
# add random numbers between 150 and 180 in column height
random_numbers = np.random.randint(150,180,size = len(df))
df['Height'] = random_numbers
df
```

```python
# replace null value in salary using median methode
median = df['Salary'].median()
df['Salary'].fillna(median, inplace=True)
df.to_excel('ABC_dataset.xlsx', index=False)
df
            
            
```

REPORT 

```python
# Count the number of employees in each team
team_counts = df['Team'].value_counts()

# Calculate percentage split
total_employees = len(df)
percentage_split = (team_counts / total_employees) * 100
team_df = pd.DataFrame({'Team':team_counts.index,'Number of employees':team_counts.values,'Percentage':percentage_split.values})
team_df.set_index('Team',inplace=True)
team_df
```

```python
counts_team_unique = df['Team'].nunique()
counts_team_unique
```

```python

plt.figure(figsize=(10,10))
plt.pie(percentage_split.values,labels=team_counts.index, autopct='%1.1f%%')
plt.title('Percentage splitting with respect to the total employees')
plt.show()
```

```python
plt.bar(team_counts.index,team_counts.values,edgecolor="black")
plt.xlabel("Team")
plt.ylabel("Number of employees")
plt.title("Number of employees in each team")
plt.xticks(rotation='vertical')
plt.figure(figsize=(10,10))
plt.show()
```

```python
#Segregatethe employees with respect to position
positions = df['Position'].unique()
employees_by_position = {position:df[df['Position'] == position] for position in positions}
for position, employees_df in employees_by_position.items():
    print(f'Employees in position "{position}":')
    print(employees_df.to_string(index=False))
    print()
```

```python
# Age grouping
age_data = df['Age']
plt.figure(figsize=(6, 4)) 
# Create a histogram
plt.hist(age_data, bins=[15, 20, 25, 30, 35,40,45,50], edgecolor='black')

# Add labels and title
plt.xlabel("Age")
plt.ylabel("Number of Employees")
plt.title("Histogram of Age Data")

# Show the plot
plt.show()
```

```python
# Salary grouping by team
team = df['Team']
salary = df['Salary']
plt.figure(figsize=(6, 4))  # Create a figure with a specific size

plt.bar(team, salary, color='blue')
plt.xticks(rotation='vertical')
plt.title('Bar Chart ')  # Add a title
plt.xlabel('Team')  # Add x-axis label
plt.ylabel('Salary')  # Add y-axis label

plt.grid(True)  # Add a grid

plt.show()  # Display the plot
```

```python
# Salary grouping by position
position = df['Position']
salary = df['Salary']
plt.figure(figsize=(6, 4))  # Create a figure with a specific size

plt.bar(position, salary, color='blue')
plt.xticks(rotation='vertical')
plt.title('Bar Chart ')  # Add a title
plt.xlabel('Position')  # Add x-axis label
plt.ylabel('Salary')  # Add y-axis label

plt.grid(True)  # Add a grid

plt.show()  # Display the plot
```

```python
# Find out under which team and position Spending in terms of salary is high
max_salary_spending =  df.groupby(['Team', 'Position'])['Salary'].sum().idxmax()
max_salary_spending
```

```python
# Correlation between Age and Salary
correlation = df['Age'].corr(df['Salary'])

print(f"The correlation between age and salary is: {correlation:.5f}")
```

```python
# Scatter plot of Age Vs Salary
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Salary'], alpha=0.4)
plt.title('Scatter Plot of Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```

```python
# Correlation between Age an Salary Visually
correlation_matrix = df[['Age', 'Salary']].corr()
print(correlation_matrix)

# Create a heatmap plot
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Age and Salary')
plt.show()
```
